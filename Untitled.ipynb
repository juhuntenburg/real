```python
import matplotlib.pyplot as plt
import seaborn as sns
import nibabel as nb
import numpy as np
```

```python
def nextpow2(i):
    n = 1
    while n < i: n *= 2
    return n
```

```python
data_dir = "/home/julia/projects/real_data/mouse_visual/CL181030fmrssouris3/processed/func/4/"
mag_file = data_dir + "mag/fft/func_moco_denoised.nii"
real_file = data_dir + "real/fft/func_moco_denoised.nii"
mask_file = data_dir + "func_mask.nii.gz"
```

```python
real = nb.load(real_file).get_data()
mag = nb.load(mag_file).get_data()
mask = nb.load(mask_file).get_data()
```

```python
paradigm = 38*[0]+24*[1]+48*[0]+24*[1]+48*[0]+24*[1]+48*[0]+24*[1]+48*[0]+24*[1]+48*[0]
```

```python
nfft = 2^nextpow2(real.shape[3])
TR=1
Fs=1/TR
f=Fs/2*np.linspace(0,1,int(nfft/2+1));
```

```python
real_fft = np.absolute(np.fft.rfft(real, nfft)/real.shape[3])
mag_fft = np.absolute(np.fft.rfft(mag, nfft)/real.shape[3])
para_fft = np.absolute(np.fft.rfft(paradigm, nfft)/real.shape[3])
```

```python
sns.set_context("poster")
plt.figure(figsize=(10,5))
plt.plot(f, para_fft, label='paradigm')
#plt.plot(real_fft[53,35,3,:], label='real')
#plt.plot(mag_fft[53,35,3,:], label='magnitude')
plt.legend()
sns.despine()
```
