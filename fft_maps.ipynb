```python
import matplotlib.pyplot as plt
import seaborn as sns
import nibabel as nb
import numpy as np
from nilearn import plotting
```

```python
dataset = 'CL181030fmrssouris3'
scan1 = '4'
scan2 = '26'
data_dir1 = "/home/julia/projects/real_data/mouse_visual/%s/processed/func/%s/"%(dataset, scan1)
data_dir2 = "/home/julia/projects/real_data/mouse_visual/%s/processed/func/%s/"%(dataset, scan2)
struct_file = "/home/julia/projects/real_data/mouse_visual/%s/processed/struct/struct_masked.nii.gz"%(dataset)
mag_file1 = data_dir1 + "mag/fft/fft2struct.nii"
real_file1 = data_dir1 + "real/fft/fft2struct.nii"
mag_file2 = data_dir2 + "mag/fft/fft2struct.nii"
real_file2 = data_dir2 + "real/fft/fft2struct.nii"
```

```python
mag1 = nb.load(mag_file1).get_data()
real1 = nb.load(real_file1).get_data()
mag2 = nb.load(mag_file2).get_data()
real2 = nb.load(real_file2).get_data()
struct = nb.load(struct_file).get_data()
```

```python
vmax = max(np.nanmax(mag1), np.nanmax(real1), np.nanmax(mag2), np.nanmax(real2))
vmin = 0

fig=plt.figure(figsize=(20,30))
slices=[6,7,8,9,10,11,12,13]
for s in range(len(slices)):        
    fig.add_subplot(8,4,(4*s)+1)
    plt.imshow(struct[:,:,slices[s]].T, cmap='Greys', vmin=np.nanmin(struct), vmax=np.nanmax(struct))
    plt.imshow(mag1[:,:,slices[s]].T, cmap='seismic', vmin=vmin, vmax=vmax, alpha=0.5)
    plt.axis('off');
    fig.add_subplot(8,4,(4*s)+2)
    plt.imshow(struct[:,:,slices[s]].T, cmap='Greys', vmin=np.nanmin(struct), vmax=np.nanmax(struct))
    plt.imshow(mag2[:,:,slices[s]].T, cmap='seismic', vmin=vmin, vmax=vmax, alpha=0.5)
    plt.axis('off');
    fig.add_subplot(8,4,(4*s)+3)
    plt.imshow(struct[:,:,slices[s]].T, cmap='Greys', vmin=np.nanmin(struct), vmax=np.nanmax(struct))
    plt.imshow(real1[:,:,slices[s]].T, cmap='seismic', vmin=vmin, vmax=vmax, alpha=0.5)
    plt.axis('off');
    fig.add_subplot(8,4,(4*s)+4)
    plt.imshow(struct[:,:,slices[s]].T, cmap='Greys', vmin=np.nanmin(struct), vmax=np.nanmax(struct))
    plt.imshow(real2[:,:,slices[s]].T, cmap='seismic', vmin=vmin, vmax=vmax, alpha=0.5)
    plt.axis('off');
```
